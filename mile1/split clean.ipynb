```python
from sklearn.model_selection import train_test_split

# Load the preprocessed dataset
clean_dataset = pd.read_csv('clean_csv.csv')

# Split the dataset into train, validate, and test sets (80%, 10%, 10%)
train, test = train_test_split(clean_dataset, test_size=0.2, random_state=42)
validate, test = train_test_split(test, test_size=0.5, random_state=42)

# Save the split datasets
train.to_csv('train1.csv', index=False)
validate.to_csv('validate1.csv', index=False)
test.to_csv('test1.csv', index=False)

print("Datasets split and saved as train1.csv, validate1.csv, and test1.csv.")
```

```python
total_records = len(clean_dataset)
train_ratio = len(train) / total_records
validate_ratio = len(validate) / total_records
test_ratio = len(test) / total_records

print("Training set proportion:", train_ratio)
print("Validation set proportion:", validate_ratio)
print("Test set proportion:", test_ratio)
```

```python
import matplotlib.pyplot as plt

# Proportions
labels = ['Train', 'Validate', 'Test']
sizes = [len(train), len(validate), len(test)]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, sizes, color=['blue', 'orange', 'green'])
plt.xlabel('Dataset')
plt.ylabel('Number of Records')
plt.title('Proportions of Train, Validate, and Test Sets')
plt.show()
```
