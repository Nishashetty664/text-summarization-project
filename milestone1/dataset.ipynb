reducing size to 11,490 and saved to demo.csv

```python
import pandas as pd  # Import pandas for data manipulation and analysis

def main():
    # Load the dataset from a CSV file
    data = pd.read_csv('/content/drive/My Drive/dataset.csv')

    # Display the original size of the dataset
    print(f"Original dataset size: {len(data)}")

    # Calculate the fraction needed to reduce the size to 11,490 records
    sample_fraction = 11490 / len(data)

    # Perform random sampling to reduce the dataset size
    sampled_data = data.sample(frac=sample_fraction, random_state=1)

    # Display the new size of the dataset
    print(f"Sampled dataset size: {len(sampled_data)}")

    # Save the sampled dataset to a new CSV file
    sampled_data.to_csv('/content/drive/My Drive/test.csv', index=False)

    print("Sampling completed. Sampled data saved to 'test.csv'.")

if __name__ == "__main__":
    # Run the main function
    main()
```

## graph to show total records in both datasets

```python

import matplotlib.pyplot as plt  # Import matplotlib for plotting

def plot_dataset_sizes():
    # Load the datasets
    original_data = pd.read_csv('/content/drive/My Drive/dataset.csv')
    reduced_data = pd.read_csv('/content/drive/My Drive/test.csv')

    # Calculate the number of records in each dataset
    original_size = len(original_data)
    reduced_size = len(reduced_data)

    # Create a bar graph to represent the dataset sizes
    plt.figure(figsize=(10, 6))  # Set the figure size
    plt.bar(['Original Dataset', 'Reduced Dataset'], [original_size, reduced_size], color=['blue', 'orange'])

    # Set the title and labels
    plt.title('Comparison of Dataset Sizes')
    plt.xlabel('Dataset')
    plt.ylabel('Number of Records')

    # Display the numbers on top of the bars
    plt.text(0, original_size, str(original_size), ha='center', va='bottom', fontsize=12)
    plt.text(1, reduced_size, str(reduced_size), ha='center', va='bottom', fontsize=12)

    # Show the plot
    plt.show()

if __name__ == "__main__":
    plot_dataset_sizes()
```
